## 📦 PASO 1: Imports y Setup Inicial

```python
import os
import sys
import pandas as pd
import numpy as np
from pathlib import Path
import json
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Para visualizaciones
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

print("✅ Imports completados")
print(f"📅 Fecha de diagnóstico: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

## 🗂️ PASO 2: Mapeo Completo del Proyecto

```python
# Define la ruta raíz de tu proyecto
# AJUSTA ESTA RUTA A TU PROYECTO
PROJECT_ROOT = Path(r"C:\Users\JUAN\DashLogistics")

print(f"📁 Directorio del proyecto: {PROJECT_ROOT}")
print(f"📁 Directorio actual: {Path.cwd()}")

# Función para mapear estructura
def map_project_structure(root_path, max_depth=3, current_depth=0):
    """
    Mapea la estructura del proyecto de forma recursiva
    """
    structure = {}
    
    if current_depth >= max_depth:
        return structure
    
    try:
        for item in sorted(root_path.iterdir()):
            # Ignorar carpetas comunes que no necesitamos
            if item.name in ['.git', '__pycache__', 'venv', '.env', 'node_modules', '.ipynb_checkpoints']:
                continue
            
            if item.is_file():
                # Guardar info del archivo
                size = item.stat().st_size
                structure[item.name] = {
                    'type': 'file',
                    'extension': item.suffix,
                    'size_kb': round(size / 1024, 2),
                    'path': str(item)
                }
            elif item.is_dir():
                # Recursión para subdirectorios
                structure[item.name] = {
                    'type': 'directory',
                    'contents': map_project_structure(item, max_depth, current_depth + 1)
                }
    except PermissionError:
        pass
    
    return structure

# Mapear la estructura
project_structure = map_project_structure(PROJECT_ROOT)

# Mostrar de forma legible
def print_structure(structure, indent=0):
    for name, info in structure.items():
        prefix = "  " * indent
        if info['type'] == 'file':
            print(f"{prefix}📄 {name} ({info['size_kb']} KB) - {info['extension']}")
        else:
            print(f"{prefix}📁 {name}/")
            if info['contents']:
                print_structure(info['contents'], indent + 1)

print("\n🌲 ESTRUCTURA DEL PROYECTO:\n")
print_structure(project_structure)
```

## 🔎 PASO 3: Inventario de Scripts y Notebooks

```python
# Buscar todos los archivos Python y notebooks
def find_files_by_extension(root_path, extensions):
    """
    Encuentra todos los archivos con extensiones específicas
    """
    files = {}
    for ext in extensions:
        files[ext] = list(root_path.rglob(f"*{ext}"))
    return files

# Buscar scripts y notebooks
project_files = find_files_by_extension(PROJECT_ROOT, ['.py', '.ipynb', '.csv', '.json', '.env'])

print("📊 INVENTARIO DE ARCHIVOS:\n")
for ext, files in project_files.items():
    print(f"\n{ext.upper()} files: {len(files)}")
    for f in files:
        rel_path = f.relative_to(PROJECT_ROOT)
        size = f.stat().st_size / 1024
        print(f"  • {rel_path} ({size:.2f} KB)")
```

## 📂 PASO 4: Verificar Estructura de Datos

```python
# Buscar carpetas de datos
data_folders = ['data', 'data/raw', 'data/clean', 'data/processed', 'data/figures']

print("📂 VERIFICACIÓN DE CARPETAS DE DATOS:\n")
for folder in data_folders:
    folder_path = PROJECT_ROOT / folder
    exists = folder_path.exists()
    status = "✅" if exists else "❌"
    print(f"{status} {folder}")
    
    if exists:
        # Listar archivos dentro
        files = list(folder_path.iterdir())
        if files:
            print(f"   Archivos encontrados: {len(files)}")
            for f in files[:5]:  # Mostrar solo los primeros 5
                if f.is_file():
                    size = f.stat().st_size / 1024
                    print(f"   • {f.name} ({size:.2f} KB)")
        else:
            print("   ⚠️ Carpeta vacía")

print("\n" + "="*60)
```

## 🔬 PASO 5: Análisis de Archivos de Datos CSV

```python
# Analizar todos los CSV encontrados
csv_files = project_files.get('.csv', [])

print(f"📊 ANÁLISIS DE {len(csv_files)} ARCHIVOS CSV:\n")

csv_analysis = {}

for csv_file in csv_files:
    try:
        print(f"\n{'='*60}")
        print(f"📄 Archivo: {csv_file.name}")
        print(f"📍 Ruta: {csv_file.relative_to(PROJECT_ROOT)}")
        
        # Intentar leer el CSV
        df = pd.read_csv(csv_file, nrows=5)  # Solo las primeras 5 filas para análisis rápido
        
        print(f"\n✅ Lectura exitosa")
        print(f"   Columnas: {len(df.columns)}")
        print(f"   Muestra (primeras 5 filas):")
        print(f"\n   Columnas: {list(df.columns)}")
        
        # Leer el archivo completo para estadísticas
        df_full = pd.read_csv(csv_file)
        print(f"\n   📏 Dimensiones completas: {df_full.shape[0]} filas × {df_full.shape[1]} columnas")
        print(f"   📊 Memoria: {df_full.memory_usage(deep=True).sum() / 1024:.2f} KB")
        
        # Tipos de datos
        print(f"\n   📋 Tipos de datos:")
        for col, dtype in df_full.dtypes.items():
            nulls = df_full[col].isnull().sum()
            null_pct = (nulls / len(df_full)) * 100
            print(f"      • {col}: {dtype} (Nulos: {nulls} / {null_pct:.1f}%)")
        
        # Guardar en análisis
        csv_analysis[csv_file.name] = {
            'path': str(csv_file.relative_to(PROJECT_ROOT)),
            'rows': df_full.shape[0],
            'columns': df_full.shape[1],
            'column_names': list(df_full.columns),
            'dtypes': df_full.dtypes.to_dict(),
            'null_counts': df_full.isnull().sum().to_dict(),
            'status': 'OK'
        }
        
    except Exception as e:
        print(f"\n❌ Error al leer: {e}")
        csv_analysis[csv_file.name] = {
            'path': str(csv_file.relative_to(PROJECT_ROOT)),
            'status': 'ERROR',
            'error': str(e)
        }

print(f"\n{'='*60}")
print(f"\n✅ Análisis de CSV completado")
```

## 🧪 PASO 6: Verificar Variables de Entorno (.env)

```python
from dotenv import load_dotenv

# Buscar archivo .env
env_file = PROJECT_ROOT / '.env'
env_example_file = PROJECT_ROOT / '.env.example'

print("🔐 VERIFICACIÓN DE VARIABLES DE ENTORNO:\n")

if env_file.exists():
    print("✅ Archivo .env encontrado")
    load_dotenv(env_file)
    
    # Leer el contenido (sin mostrar valores sensibles)
    with open(env_file, 'r') as f:
        lines = f.readlines()
        print(f"\n   Variables configuradas: {len([l for l in lines if '=' in l and not l.startswith('#')])}")
        print("\n   Variables encontradas:")
        for line in lines:
            if '=' in line and not line.startswith('#'):
                var_name = line.split('=')[0].strip()
                print(f"      • {var_name}")
else:
    print("❌ Archivo .env NO encontrado")

if env_example_file.exists():
    print("\n✅ Archivo .env.example encontrado")
    with open(env_example_file, 'r') as f:
        lines = f.readlines()
        print(f"   Variables de ejemplo: {len([l for l in lines if '=' in l and not l.startswith('#')])}")
else:
    print("\n⚠️ Archivo .env.example NO encontrado")

# Verificar variables clave
print("\n🔍 Verificando variables críticas del sistema:")
critical_vars = ['RAW_DATA_PATH', 'CLEAN_DATA_PATH', 'DATA_URL', 'WEATHER_API_KEY']

for var in critical_vars:
    value = os.getenv(var)
    if value:
        # Ocultar parte del valor si es sensible
        if 'KEY' in var or 'PASSWORD' in var or 'TOKEN' in var:
            display_value = value[:5] + '...' + value[-5:] if len(value) > 10 else '***'
        else:
            display_value = value
        print(f"   ✅ {var} = {display_value}")
    else:
        print(f"   ❌ {var} = NO CONFIGURADA")
```

## 🔧 PASO 7: Probar Módulos del Pipeline

```python
# Añadir el directorio src al path para importar módulos
src_path = PROJECT_ROOT / 'src'
if src_path.exists():
    sys.path.insert(0, str(src_path))
    print(f"✅ Directorio src añadido al path: {src_path}")
else:
    print(f"❌ Directorio src NO encontrado en {PROJECT_ROOT}")

# Buscar módulos Python en src
print("\n🐍 MÓDULOS PYTHON ENCONTRADOS EN SRC:\n")

if src_path.exists():
    python_files = list(src_path.glob('*.py'))
    
    for py_file in python_files:
        if py_file.name == '__init__.py':
            continue
        
        print(f"\n📄 {py_file.name}")
        
        # Intentar importar
        module_name = py_file.stem
        try:
            module = __import__(module_name)
            print(f"   ✅ Importación exitosa")
            
            # Listar funciones/clases principales
            members = [m for m in dir(module) if not m.startswith('_')]
            if members:
                print(f"   📋 Funciones/clases disponibles: {', '.join(members[:5])}")
                if len(members) > 5:
                    print(f"      ... y {len(members) - 5} más")
        except Exception as e:
            print(f"   ❌ Error al importar: {e}")
else:
    print("⚠️ No se puede verificar módulos - directorio src no existe")
```

## 🌦️ PASO 8: Diagnóstico de Datos del Clima

```python
print("🌦️ DIAGNÓSTICO DE DATOS METEOROLÓGICOS:\n")

# Verificar si hay algún CSV o JSON con datos del clima
weather_keywords = ['weather', 'clima', 'temperature', 'temp', 'forecast']

weather_files = []
for ext in ['.csv', '.json']:
    for file in project_files.get(ext, []):
        if any(keyword in file.name.lower() for keyword in weather_keywords):
            weather_files.append(file)

if weather_files:
    print(f"✅ Archivos de clima encontrados: {len(weather_files)}\n")
    for wf in weather_files:
        print(f"   📄 {wf.relative_to(PROJECT_ROOT)}")
        try:
            if wf.suffix == '.csv':
                df_weather = pd.read_csv(wf)
                print(f"      Dimensiones: {df_weather.shape}")
                print(f"      Columnas: {list(df_weather.columns)}")
            elif wf.suffix == '.json':
                with open(wf, 'r') as f:
                    data = json.load(f)
                print(f"      Estructura JSON: {type(data)}")
                if isinstance(data, list):
                    print(f"      Registros: {len(data)}")
        except Exception as e:
            print(f"      ❌ Error al leer: {e}")
else:
    print("❌ NO se encontraron archivos de datos del clima")
    print("\n💡 POSIBLES CAUSAS:")
    print("   • No se ha ejecutado el script de descarga de clima")
    print("   • La API key del clima no está configurada")
    print("   • Los archivos están en otra ubicación")
    print("   • El pipeline de clima tiene errores")

# Verificar API key del clima
weather_api_key = os.getenv('WEATHER_API_KEY') or os.getenv('OPENWEATHER_API_KEY')
if weather_api_key:
    print(f"\n✅ API Key de clima configurada: {weather_api_key[:10]}...")
else:
    print("\n❌ API Key de clima NO configurada")
    print("   Necesitas obtener una API key de un servicio como OpenWeatherMap")
```

## 🎯 PASO 9: Verificar Dashboard Streamlit

```python
print("🎯 VERIFICACIÓN DEL DASHBOARD STREAMLIT:\n")

# Buscar archivos de Streamlit
streamlit_files = [f for f in project_files.get('.py', []) if 'streamlit' in f.name.lower() or 'dashboard' in f.name.lower() or 'app' in f.name.lower()]

if streamlit_files:
    print(f"✅ Archivos de dashboard encontrados: {len(streamlit_files)}\n")
    
    for st_file in streamlit_files:
        print(f"📄 {st_file.relative_to(PROJECT_ROOT)}")
        
        # Leer el contenido y buscar qué datos carga
        try:
            with open(st_file, 'r', encoding='utf-8') as f:
                content = f.read()
                
            # Buscar referencias a archivos de datos
            data_references = []
            for line in content.split('\n'):
                if 'pd.read_csv' in line or 'pd.read_json' in line:
                    data_references.append(line.strip())
            
            if data_references:
                print("\n   📊 Datos que intenta cargar:")
                for ref in data_references[:10]:  # Mostrar max 10
                    print(f"      • {ref}")
            
            # Buscar imports
            imports = [line for line in content.split('\n') if line.strip().startswith('import ') or line.strip().startswith('from ')]
            print(f"\n   📦 Dependencias: {len(imports)}")
            
        except Exception as e:
            print(f"   ❌ Error al analizar: {e}")
else:
    print("❌ NO se encontraron archivos de dashboard Streamlit")
    print("\n💡 Busca archivos como:")
    print("   • app.py")
    print("   • dashboard.py")
    print("   • streamlit_app.py")
```
